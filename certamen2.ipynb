```python
import pandas as pd
from functools import partial
from pandas.tseries.frequencies import to_offset


df1 = pd.DataFrame({
    'fecha': pd.date_range('2024-01-01 00:00:00', periods=10, freq='30T'),
    'valor': range(10)
})


def roundfunc(t, freq):
    freq = to_offset(freq)
    return pd.Timestamp((t.value // freq.delta.value) * freq.delta.value)


grouped = df1.groupby(df1['fecha'].apply(partial(roundfunc, freq='120T')))


attrs = {
    'var': 'grouped',
    'type': type(grouped).__name__,
}
print(attrs)


enum = enumerate(grouped.groups)
nextel = next(enum)


grupo_resultado = grouped.get_group(nextel[1])
print(grupo_resultado)
```

```python
import pandas as pd
from functools import partial
from pandas.tseries.frequencies import to_offset


df1 = pd.DataFrame({
    'fecha': pd.date_range('2024-01-01 00:00:00', periods=10, freq='30T'),
    'valor': range(10)
})


df1['fecha'] = pd.to_datetime(df1['fecha'])


def roundfunc(t, freq):
    freq = to_offset(freq)
    return pd.Timestamp((t.value // freq.delta.value) * freq.delta.value)


df1['fecha_redondeada'] = df1['fecha'].apply(partial(roundfunc, freq='120T'))


grouped = df1.groupby('fecha_redondeada')


df_median = grouped.aggregate('median')


attrs = {
    'var': 'df_median',
    'type': type(df_median).__name__,
    'shape': df_median.shape,
    'columns': list(df_median.columns),
}


print(attrs)
print(df_median)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
import logging


logging.basicConfig(level=logging.INFO, format='%(levelname)s: %(message)s')


def encode_categorical_columns(df: pd.DataFrame) -> pd.DataFrame:
    df_encoded = df.copy()
    cat_cols = df_encoded.select_dtypes(include=['category']).columns
    for col in cat_cols:
        df_encoded[col] = df_encoded[col].cat.codes
    return df_encoded


def create_datetime_windows(dfx, dfy, dfy_total, size, periods, y_choice='last', ahead=''):
    x_windowed = []
    y_windowed = []
    window_indices = []

    if not isinstance(dfx.index, pd.DatetimeIndex) or not isinstance(dfy.index, pd.DatetimeIndex):
        raise TypeError("Los DataFrames deben tener índices de tipo DatetimeIndex.")

    def _process_window(window_series):
        current_window_indices = window_series.index
        if len(current_window_indices) != periods:
            return None

        if y_choice == 'ahead':
            try:
                future_index = current_window_indices[-1] + pd.Timedelta(ahead)
                wy = dfy_total.loc[[future_index]]
                if wy.empty:
                    return None
                wx = dfx.loc[current_window_indices]
                x_windowed.append(wx)
                y_windowed.append(wy)
                window_indices.append(current_window_indices[0])
            except Exception:
                return None
        else:
            wx = dfx.loc[current_window_indices]
            wy = dfy.loc[current_window_indices]
            if y_choice == 'last':
                wy = wy.iloc[-1]
            elif y_choice == 'first':
                wy = wy.iloc[0]
            x_windowed.append(wx)
            y_windowed.append(wy)
            window_indices.append(current_window_indices[0])
        return 0

    dfx.iloc[:, 0].rolling(window=size, min_periods=periods).apply(lambda w: _process_window(w), raw=False)
    return x_windowed, y_windowed, window_indices


def get_list_shape_string(data_list: list) -> str:
    if not data_list:
        return 'empty'
    first_element = data_list[0]
    if isinstance(first_element, (np.ndarray, pd.DataFrame, pd.Series)):
        shape_str = str(first_element.shape)
    else:
        try:
            shape_str = str(np.array(first_element).shape)
        except Exception:
            shape_str = "unknown_shape"
    return f"{len(data_list)} x {shape_str}"


def normal_anomal_split(dfx, dfy, size, periods, y_choice='last', ahead='', custom_normal_split=False, normal_test_ratio=0.1, scaler_type='MinMaxScaler'):
    scaler_map = {
        'MinMaxScaler': MinMaxScaler(),
        'StandardScaler': StandardScaler(),
        'RobustScaler': RobustScaler()
    }
    myscaler = scaler_map.get(scaler_type)
    if myscaler is None:
        raise ValueError(f"Escalador no válido: {scaler_type}")

    dfx_encoded = encode_categorical_columns(dfx)
    dfy_encoded = encode_categorical_columns(dfy)

    x_windowed, y_windowed, windows_ind = create_datetime_windows(dfx_encoded, dfy_encoded, dfy_encoded, size, periods, y_choice, ahead)
    if not x_windowed:
        return {'x_train': np.array([]), 'x_test': np.array([]), 'y_test': np.array([]), 'x_mix': np.array([]), 'y_mix': np.array([]), 'i_mix': [], 'myscaler': myscaler}

    x_normal, y_normal = [], []
    x_anomal, y_anomal = [], []

    for i, wy in enumerate(y_windowed):
        label = wy.iloc[0].item() if isinstance(wy, (pd.Series, pd.DataFrame)) else wy
        if label == 0:
            x_normal.append(x_windowed[i])
            y_normal.append(y_windowed[i])
        else:
            x_anomal.append(x_windowed[i])
            y_anomal.append(y_windowed[i])

    normal_indices = np.arange(len(x_normal))
    np.random.shuffle(normal_indices)

    if custom_normal_split:
        length_test_normal = round(len(x_normal) * normal_test_ratio)
    else:
        length_test_normal = min(len(x_anomal), len(x_normal) - 1)

    normal_test_indices = normal_indices[:length_test_normal]
    normal_train_indices = normal_indices[length_test_normal:]

    x_normal_train = [x_normal[i] for i in normal_train_indices]
    x_normal_test = [x_normal[i] for i in normal_test_indices]

    x_normal_train_filtered = [df for df in x_normal_train if not df.empty]
    x_normal_train_total_flat = pd.concat(x_normal_train_filtered).to_numpy() if x_normal_train_filtered else np.array([])

    is_scaler_fitted = False
    if x_normal_train_total_flat.size > 0:
        try:
            myscaler.fit(x_normal_train_total_flat)
            is_scaler_fitted = True
        except Exception:
            pass

    def _scale_and_squeeze(data_list, scaler_obj):
        scaled_list = []
        for item_df in data_list:
            if item_df.empty or not hasattr(scaler_obj, 'scale_'):
                scaled_list.append(np.array([]))
                continue
            scaled_array = scaler_obj.transform(item_df.to_numpy())
            if scaled_array.ndim > 1 and scaled_array.shape[1] == 1:
                scaled_list.append(np.squeeze(scaled_array, axis=1))
            else:
                scaled_list.append(scaled_array)
        return scaled_list

    x_train_scaled = _scale_and_squeeze(x_normal_train, myscaler) if is_scaler_fitted else []
    x_test_normal_scaled = _scale_and_squeeze(x_normal_test, myscaler) if is_scaler_fitted else []
    x_anomal_scaled = _scale_and_squeeze(x_anomal, myscaler) if is_scaler_fitted else []
    x_mix_scaled = _scale_and_squeeze(x_windowed, myscaler) if is_scaler_fitted else []

    y_test = [0] * len(x_test_normal_scaled) + [1] * len(x_anomal_scaled)
    x_test = x_test_normal_scaled + x_anomal_scaled

    y_mix = [wy.iloc[0].item() if isinstance(wy, (pd.Series, pd.DataFrame)) else wy for wy in y_windowed]

    return {
        'x_train': np.array(x_train_scaled, dtype=object),
        'x_test': np.array(x_test, dtype=object),
        'y_test': np.array(y_test),
        'x_mix': np.array(x_mix_scaled, dtype=object),
        'y_mix': np.array(y_mix),
        'i_mix': windows_ind,
        'myscaler': myscaler,
    }




dates = pd.date_range(start='2023-01-01', periods=1000, freq='10T')


data_x = {
    'feature1': np.random.rand(1000),
    'feature2': np.random.randint(0, 5, 1000),
    'category_col': pd.Series(np.random.choice(['A', 'B', 'C'], 1000), dtype='category')
}
df15_features = pd.DataFrame(data_x, index=dates)


labels = np.zeros(1000)
labels[np.random.choice(1000, 50, replace=False)] = 1
data_y = {
    'Estado': pd.Series(labels, dtype='category'),
    'other_info': np.random.rand(1000)
}
df16_labels = pd.DataFrame(data_y, index=dates)


results = normal_anomal_split(
    dfx=df15_features,
    dfy=df16_labels,
    size='600T',
    periods=60,
    y_choice='last',
    custom_normal_split=True,
    normal_test_ratio=0.2,
    scaler_type='MinMaxScaler'
)


print("--- Resultados ---")
print("x_train shape:", results['x_train'].shape)
print("x_test shape:", results['x_test'].shape)
print("y_test shape:", results['y_test'].shape)
print("y_mix:", results['y_mix'][:5])
print("Índices de ventanas:", results['i_mix'][:5])
```
